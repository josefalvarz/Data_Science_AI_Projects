```python
# Import relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import os
```

```python
print(os.listdir())
```

```python
# Save and read the dataframe
data = pd.read_csv('Downloads/1.01.+Simple+linear+regression (1).csv')
```

```python
data
```

```python
#Descriptive Statistics of the dataset
data.describe()
```

```python
#Create input and target
y = data['GPA']
x1 = data['SAT']
```

```python
#Plot the data in a graph to see for potential patterns
plt.scatter(x1,y)
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
#Train the model
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
#Plot the scatter to see the relationships between the two variables.
plt.scatter(x1,y)
yhat = 0.0017*x1+0.275
fig = plt.plot(x1,yhat,lw=5,c='orange',label = 'regression line')
plt.xlabel('SAT',fontsize = 20)
plt.ylabel('GPA',fontsize = 20)
plt.show
```
