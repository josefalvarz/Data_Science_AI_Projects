```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('1.01.+Simple+linear+regression (2).csv')
data.head() #View first 5 rows
```

```python
x = data['SAT']
y = data['GPA']
```

```python
x.shape
```

```python
y.shape
```

```python
x_matrix = x.values.reshape(-1,1)
x_matrix.shape
```

```python
reg = LinearRegression()
reg.fit(x_matrix,y)
```

```python
reg.score(x_matrix,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[1203]]) # We trained the model as a 2D so scikit learn expects a 2D array, that's the reason of the brackets
```

```python
new_data = pd.DataFrame(data = [[1203],[1740]],columns = ['SAT'])
new_data
```

```python
reg.predict(new_data[['SAT']])
```

```python
new_data['Predicted GPA'] = reg.predict(new_data[['SAT']])
new_data
```
