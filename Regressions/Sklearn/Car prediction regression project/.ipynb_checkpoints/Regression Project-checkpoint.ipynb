```python
#Import the relevant libraries
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
sns.set()
```

```python
#Uploaad the data
raw_data = pd.read_csv('1.04.+Real-life+example.csv')
raw_data.head()
```

```python
#Evaluate and identify potential issues in our data using descriptive statistics
raw_data.describe(include='all')
```

```python
#Remove unnecessary features
data = raw_data.drop(['Model'],axis=1)
data
```

```python
#Count how many missing values are per feature
data.isnull().sum()
```

```python
#We remove the missing values from Price and EngineV because this missing values are less than the 5% of each feature data
data_No_MV = data.dropna(axis=0)
data_No_MV.describe(include='all')
```

```python
#Probality Distribution Formula
sns.displot(data_No_MV['Price']) ## Displays a graph containing the distribution of each feature we ask for.
```

```python
#Dealing with outliers, use of quantile to remove outliers
quantile = data_No_MV['Price'].quantile(0.99)
data_1 = data_No_MV[data_No_MV['Price'] < quantile]
data_1.describe()
```

```python
sns.displot(data_1['Price'])
```

```python
sns.displot(data_1['Mileage'])
```

```python
quantile2 = data_1['Mileage'].quantile(0.99)
data_2 = data_1[data_1['Mileage']<quantile2]
```

```python
sns.displot(data_2['Mileage'])
```

```python
sns.displot(data_2['EngineV'])
```

```python
data_3 = data_2[data_2['EngineV']<6.5] ## Some values for engine volume were wrong, they usally range between 0.6 and 6.5
sns.displot(data_3['EngineV'])
```

```python
sns.displot(data_3['Year'])
```

```python
quantile3 = data_3['Year'].quantile(0.01) #Removing the outliers for year
data_4 = data_3[data_3['Year'] > quantile3]
data_4.describe()
```

```python
sns.displot(data_4['Year'])
```

```python
#Creating a new index and delete the older one
data_cleaned = data_4.reset_index(drop=True)
data_cleaned.describe(include='all')
```

```python
#Checking OLS assumptions
#Scatter plot to check for linearity
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Price'])
ax1.set_title('Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Price'])
ax2.set_title('Price and Engine V')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Price'])
ax3.set_title('Price and Mileage')

plt.show()
#The graph shows clear patterns but is not linear, we should transform one or more features
```

```python
sns.displot(data_cleaned['Price'])
#The distribution of price is a reason of why the plots above are not linear, we should apply the log transformation to 'Price'.
```

```python
log_price = np.log(data_cleaned['Price'])
data_cleaned['Log price'] = log_price
data_cleaned
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(data_cleaned['Year'],data_cleaned['Log price'])
ax1.set_title('Log Price and Year')
ax2.scatter(data_cleaned['EngineV'],data_cleaned['Log price'])
ax2.set_title('Log Price and Engine V')
ax3.scatter(data_cleaned['Mileage'],data_cleaned['Log price'])
ax3.set_title('Log Price and Mileage')
```

```python
data_cleaned = data_cleaned.drop(['Price'],axis=1)
data_cleaned
```

```python
#Check for multicollinearity
data_cleaned.columns.values
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_cleaned[['Mileage','Year','EngineV']]
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['features'] = variables.columns
```

```python
vif
```

```python
#Year demonstrates that has a high degree of multicollinearity in comparison to the other features, therefore we should remove it
data_no_multicollinearity = data_cleaned.drop(['Year'],axis=1)
data_no_multicollinearity
```

```python
#We create the dummies variables for each feature, ensure drop_first is set it to 1 because this way we are having for each category of dummies n-1.
#Where n = total number of categories a categorical variable has.
data_with_dummies = pd.get_dummies(data_no_multicollinearity,drop_first=1,dtype=float)
```

```python
data_with_dummies
```

```python
#Rearrange the dataset
data_with_dummies.columns.values
```

```python
cols = ['Log price', 'Mileage', 'EngineV', 'Brand_BMW',
       'Brand_Mercedes-Benz', 'Brand_Mitsubishi', 'Brand_Renault',
       'Brand_Toyota', 'Brand_Volkswagen', 'Body_hatch', 'Body_other',
       'Body_sedan', 'Body_vagon', 'Body_van', 'Engine Type_Gas',
       'Engine Type_Other', 'Engine Type_Petrol', 'Registration_yes']
```

```python
data_preprocessed = data_with_dummies[cols]
data_preprocessed
```

```python
##Variance inflation factor excluding the logprice
variables = data_preprocessed.drop(['Log price'],axis=1)
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif['Features'] = variables.columns
vif
```

```python
#Declare the inputs and outputs of the regression
targets = data_preprocessed['Log price']
inputs = data_preprocessed.drop(['Log price'],axis=1)
```

```python
#Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(inputs)
x_scaled = scaler.transform(inputs)
```

```python
#Split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, targets, test_size=0.20, random_state = 43)
```

```python
#Create the regression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#Compare the predicted outcomes with the target values
y_hat = reg.predict(x_train)
plt.scatter(y_train,y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
plt.xlim(6,13)
plt.xlim(6,13)
plt.show()
```

```python
#Plot the residuals
y_hat = reg.predict(x_train)
sns.displot(y_train-y_hat)
plt.title('Residuals PDF',size=18)
```

```python
#Find the R-squared
reg.score(x_train,y_train)
```

```python
#Find the bias
reg.intercept_
```

```python
#Find the weights
reg.coef_
```

```python
weights_table = pd.DataFrame(inputs.columns.values, columns=[['Features']])
weights_table['Weights'] = reg.coef_
weights_table
```

```python
##Testing
#Plot the target values against the predicted values
y_hat_test = reg.predict(x_test)
plt.scatter(y_test,y_hat_test,alpha=0.2) #alpha = use for visualization to see where the higher concentration of points is based on their opacity.
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.xlim(6,13)
plt.show()
#As shown our model performs when predicting higher prices due to the large concentration of high prices, but lower prices are more scattered which
#demonstrates that our model do not perform well on predicting lower prices.
```

```python
#Creation of dataframe to visally see the differences between targets and predictions
df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Predictions'])
df_pf.head()
```

```python
y_test = y_test.reset_index(drop=True)
df_pf['Targets'] = np.exp(y_test)
df_pf
```

```python
df_pf['Residuals'] = df_pf['Targets']-df_pf['Predictions']
df_pf['Difference%'] = np.absolute(df_pf['Residuals']/df_pf['Targets']*100)
df_pf
```

```python
#Analyze the model results
df_pf.describe()
```
